# Section 4: Drawing Circles and Semicircles

Comment out the previous code (but keep `from graphics import *`). 

You can also draw filled shapes (such as circles and rectangles) with calls such as:

```python
from graphics import *

win = GraphWin("Hello", 400, 400)

# draw a green-filled circle with center (100,200)
# and radius 50
circ = Circle(Point(300, 270), 50)
circ.setFill("green")
circ.draw(win)

# draw a rectangle with opposite corners at (10,50) and (145, 280)
rect = Rectangle(Point(10,50), Point(145, 280))
rect.draw(win)
```
